

# Basic Residue - Machine Learning Models for Regression Analysis

This notebook demonstrates the application of machine learning techniques to analyze and predict values from different datasets under the Basic Residue project. The primary objective of this analysis is to evaluate the performance of linear regression models trained on various datasets and compare the models' ability to predict the target variable based on the actual observed values. The datasets used include:

- `dataset_high_scatter`
- `dataset_low_scatter`
- `dataset_negative_deviation`
- `dataset_positive_deviation`

For each dataset, the following steps are performed:

1. **Data Preprocessing**: Loading datasets, checking for missing values, and reviewing the data structure.
2. **Model Training**: Linear regression models are trained to predict the relationship between the target variable ("Predicted") and the feature ("Actual").
3. **Model Evaluation**: Performance is evaluated using metrics like Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared (RÂ²).
4. **Visualization**: Predicted vs Actual values, and residuals (Predicted - Actual) are plotted for each dataset to visually evaluate model performance.

***Section for dataset_high_scatter.***

```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#load the dataset_high_scatter
dataset_high_scatter = pd.read_csv('drive/MyDrive/Chem_ML_projects/BasicResidue/dataset_high_scatter.csv')
```

```python
#Display first 5rows of the dataset
print("First 5 rows of dataset_high_scatter.csv:")
print(dataset_high_scatter.head())
```

***Data Cleaning:***

```python
#check for missing values and errors
print("Missing values in dataset_high_scatter.csv:")
print(dataset_high_scatter.isnull().sum())
```

***Visualization:***

1.Predicted vs Actual values plot.

2.Residuals plot (Difference between predicted and actual values).

```python
# Plot Predicted vs Actual
plt.figure(figsize=(10, 6))
plt.scatter(dataset_high_scatter['Actual'], dataset_high_scatter['Predicted'], alpha=0.5, color='blue')
plt.plot([dataset_high_scatter['Actual'].min(), dataset_high_scatter['Actual'].max()],
         [dataset_high_scatter['Actual'].min(), dataset_high_scatter['Actual'].max()],
         color='red', linestyle='--')  # 45-degree line
plt.title("Predicted vs Actual Values")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()

# Calculate residuals (Predicted - Actual)
residuals = dataset_high_scatter['Predicted'] - dataset_high_scatter['Actual']

# Plot Residuals
plt.figure(figsize=(10, 6))
plt.scatter(dataset_high_scatter['Actual'], residuals, alpha=0.5, color='green')
plt.axhline(0, color='red', linestyle='--')  # Zero line
plt.title("Residuals (Predicted - Actual)")
plt.xlabel("Actual")
plt.ylabel("Residuals")
plt.show()
```

***Model Training:***


```python
X = dataset_high_scatter[['Actual']].values  # Features
y = dataset_high_scatter['Predicted'].values  # Target variable
```

```python
#Weâll use Linear Regression as an example since this is a regression problem.
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
#predict using the model
y_pred = model.predict(X)
```

***Calculate Evaluation metrics.***

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Print the evaluation metrics
print(f"Model Mean Absolute Error (MAE): {mae:.2f}")
print(f"Model Mean Squared Error (MSE): {mse:.2f}")
print(f"Model R-squared (RÂ²): {r2:.2f}")
```

***Section for dataset_low_scatter:***

```python
# Load the dataset
dataset_low_scatter = pd.read_csv('drive/MyDrive/Chem_ML_projects/BasicResidue/dataset_low_scatter.csv')
print("First 5 rows of dataset_low_scatter:")
print(dataset_low_scatter.head())
```

```python
# check for missing values and errors
print("Missing values in dataset_low_scatter.csv:")
print(dataset_low_scatter.isnull().sum())
```

***Model Training.***

```python
# Define features (X) and target (y)
X = dataset_low_scatter[['Actual']]  # Feature (Actual values)
y = dataset_low_scatter['Predicted']  # Target (Predicted values)

# Initialize and train the model (using Linear Regression as an example)
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate the model
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Print evaluation results
print(f"Model Mean Absolute Error (MAE): {mae:.2f}")
print(f"Model Mean Squared Error (MSE): {mse:.2f}")
print(f"Model R-squared (RÂ²): {r2:.2f}")
```

```python
# Plot Predicted vs Actual
plt.figure(figsize=(10, 6))
plt.scatter(dataset_low_scatter['Actual'], dataset_low_scatter['Predicted'], alpha=0.5, color='blue')
plt.plot([dataset_low_scatter['Actual'].min(), dataset_low_scatter['Actual'].max()],
         [dataset_low_scatter['Actual'].min(), dataset_low_scatter['Actual'].max()], color='red', linestyle='--')  # 45-degree line
plt.title("Predicted vs Actual Values (Low Scatter)")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()

# Calculate residuals (Predicted - Actual)
residuals_low = dataset_low_scatter['Predicted'] - dataset_low_scatter['Actual']

# Plot Residuals
plt.figure(figsize=(10, 6))
plt.scatter(dataset_low_scatter['Actual'], residuals_low, alpha=0.5, color='green')
plt.axhline(0, color='red', linestyle='--')  # Zero line
plt.title("Residuals (Predicted - Actual) (Low Scatter)")
plt.xlabel("Actual")
plt.ylabel("Residuals")
plt.show()
```
